```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_regression
```

```python
housing = pd.read_csv('housing.csv')
```

```python
housing.head()
```

```python
housing.dropna()
```

```python
# Drop rows with missing values
housing.dropna(inplace=True)
y=housing['median_house_value']
```

```python
x=housing.drop('median_house_value',axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test=\
train_test_split(x,y,test_size=0.2, random_state=22)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Identify the categorical feature
categorical_features = ['ocean_proximity']

# Create a column transformer for one-hot encoding
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",
                                  one_hot,
                                  categorical_features)],
                               remainder="passthrough")

# Apply the transformation to the training and testing data
x_train_encoded = transformer.fit_transform(x_train)
x_test_encoded = transformer.transform(x_test)

```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train_encoded, y_train)
pred = model.predict(x_test_encoded)
```

```python
# evaluate the model
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, pred)
mse
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, pred)
r2
```

```python
from sklearn.linear_model import Ridge, Lasso
model_ridge = Ridge()
model_ridge.fit(x_train_encoded, y_train)
pred_ridge = model_ridge.predict(x_test_encoded)
mse = mean_squared_error(y_test, pred_ridge)
r2 = r2_score(y_test, pred_ridge)
print(f'MSE: {mse}, R2: {r2}')
print(f'Ridge Coefficients: {model_ridge.coef_}, Intercept: {model_ridge.intercept_}')
```

```python
model_lasso = Lasso()
model_lasso.fit(x_train_encoded, y_train)
pred_lasso = model_lasso.predict(x_test_encoded)
mse = mean_squared_error(y_test, pred_lasso)
r2 = r2_score(y_test, pred_lasso)
print(f'MSE: {mse}, R2: {r2}')
print(f'Lasso Coefficients: {model_lasso.coef_}, Intercept: {model_lasso.intercept_}')
```
